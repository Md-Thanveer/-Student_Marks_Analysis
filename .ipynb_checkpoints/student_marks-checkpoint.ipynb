```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
```

```python
df = pd.read_csv('student_marks.csv')
```

```python
# Display the dataset
```

```python
print("Dataset:")
```

```python
print(df)
```

```python
Total Marks
```

```python
df['Total Marks'] = df.iloc[:, 1:].sum(axis=1)
```

```python
Average Marks
```

```python
df['Average Marks'] = df.iloc[:, 1:].mean(axis=1)
```

```python
Display the Total_ Average Marks
```

```python
print("\nTotal and Average Marks:")

print(df[['Name', 'Total Marks', 'Average Marks']])
```

```python
# 2. Identify top performers in each subject
```

```python

top_performers = {}
for subject in df.columns[1:-2]:
    top_student = df.loc[df[subject].idxmax(), 'Name']
    top_score = df[subject].max()
    top_performers[subject] = (top_student, top_score)
```

```python
print("\nTop Performers in Each Subject:")
for subject, (student, score) in top_performers.items():
    print(f"{subject}: {student} with {score} marks")
```

```python
# 3. Visualize subject-wise performance using bar charts
```

```python
subject_means = df.iloc[:, 1:-2].mean()
```

```python
plt.figure(figsize=(10, 6))
subject_means.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Marks by Subject', fontsize=16)
plt.xlabel('Subjects', fontsize=14)
plt.ylabel('Average Marks', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
 Save results to a new CSV file
```

```python
# Save results to a new CSV file
df.to_csv('student_marks_with_totals.csv', index=False)
print("\nResults saved to 'student_marks_with_totals.csv'")
```

```python
# Subject vise stats 

subject_stats = df.iloc[:, 1:-2].agg(['min', 'max'])
print("\nSubject-wise Performance (Min and Max Marks):")
print(subject_stats)
```

```python
correlation = df.iloc[:, 1:-2].corr()
print("\nCorrelation Between Subjects:")
print(correlation)
```

```python
OVERALL STUDENT RANKING 
```

```python
df['Rank'] = df['Total Marks'].rank(ascending=False, method='min')
print("\nStudent Rankings (Based on Total Marks):")
print(df[['Name', 'Total Marks', 'Rank']])
```

```python
STUDENT - WISE PERFORMANCE ACROSS SUBJECTS 
```

```python
df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set2')
plt.title('Student-wise Performance Across Subjects', fontsize=16)
plt.xlabel('Students', fontsize=14)
plt.ylabel('Marks', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```
